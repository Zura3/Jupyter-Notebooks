```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
USAhousing=pd.read_csv(r'C:\Users\rohan\Downloads\USA_Housing.csv')
```

```python
USAhousing.head()
```

```python
X=(USAhousing['Avg. Area Income'])
Y=(USAhousing['Price'])
M=len(X)
```

```python
def costFunction(theta0,theta1):
    cost=(1/(2*M))*(((theta0+theta1*X-Y)**2).sum())
    return cost
```

```python
theta0,theta1,alpha=0,0,0.00000000001
ar=[]
a=1
index=[]
costArray=[]
while(a<1000):
    theta_0=theta0-alpha*(1/M)*((theta0+theta1*X-Y).sum())
    theta_1=theta1-alpha*(1/M)*(((theta0+theta1*X-Y)*X).sum())
    if(costFunction(theta0,theta1)<=costFunction(theta_0,theta_1)):
        break
    else:
        theta0=theta_0
        theta1=theta_1
        index.append(a)
        a=a+1
        costArray.append(costFunction(theta0,theta1))
for i in range(M):
    ar.append(theta0+(theta1*X[i]))
```

```python
sns.lineplot(index,costArray)
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(X,ar,color='Red')
sns.scatterplot(X,Y)
```
