```python
plt.figure(figsize=(20,10))
sns.scatterplot(hsg['Avg. Area Income'],hsg['Price'])
```

```python
x=list(hsg['Avg. Area Income'])
y=list(hsg['Price'])
m=np.linspace(-100,100,100)
n=[]
for i in m:
    k=0
    for j in range(len(x)):
        k+=(((x[j]*i)-y[j])**2)/(len(x)*2)
    n.append(k)
```

```python
sns.scatterplot(m,n)
```

```python
n.index(min(n))
theta=(m[58])
y1=[]
z=0
for i in range(len(x)):
    z=theta*x[i]
    y1.append(z)
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x,y1,color='red')
sns.scatterplot(hsg['Avg. Area Income'],hsg['Price'])
```
