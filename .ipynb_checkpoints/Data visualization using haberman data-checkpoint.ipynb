```python
import pandas as pd
import seaborn as sns
#Importing python seabron library for data visualization
```

```python
hbn=pd.read_csv(r"D:\Data_Science\Datasets\haberman.csv",names=['age','year_of_op.','+ axillary nodes','survival status'])
#importing the haberman dataset in form of a dataframe name hbn
#naming the columns in the dataframe since they are not labelleled in the csv file
```

```python
hbn.head()
#printing the first 5 rows of the dataframe
```

```python
sns.pairplot(hbn)
#pairplot is used to plot the data for a pair of two columns
```

```python
sns.distplot(hbn['age'])
# distplot to show a univariate distribution of age.
```

```python
sns.distplot(hbn['year_of_op.'])
# distplot to show a univariate distribution of year of operation of the patients.
```

```python
sns.distplot(hbn['+ axillary nodes'])
# distplot to show a univariate distribution of the number of positive auxillary nodes.
```

```python
sns.distplot(hbn['survival status'])
# distplot to show a univariate distribution of survival status of patients
```

```python
sns.boxplot(x=hbn["age"])
#boxplot used to show distribution of frequency of age of patients.
#the box shows the quartiles of data.
```

```python
sns.boxplot(x=hbn["+ axillary nodes"])
#shows distribution of positive auxillary nodes.
```

```python
sns.violinplot(hbn["age"])
#shows density of data.
#density is max at 50-60.
```
