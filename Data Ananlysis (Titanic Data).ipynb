```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
trn=pd.read_csv(r'C:\Users\rohan\Downloads\train.csv')
```

```python
corr=trn.corr()
plt.figure(figsize=(12,12))
sns.heatmap(corr,annot=True)
```

```python
trn.isnull().sum()
```

```python
trn=trn.drop(columns=['Cabin'])
```

```python
trn['Age'].fillna(trn['Age'].mean(),inplace=True)
```

```python
train_set,test_set=train_test_split(trn,test_size=0.2)
```

```python
trainset_survived=train_set[train_set['Survived']==1]
sns.countplot(x="Sex", data=trainset_survived)
plt.plot()
```

```python
testset_survived=test_set[test_set['Survived']==1]
sns.countplot(x="Sex", data=testset_survived)
plt.plot()
```

```python
male_train=len(trainset_survived[trainset_survived['Sex']=='male'])
female_train=len(trainset_survived[trainset_survived['Sex']=='female'])
male_test=len(testset_survived[testset_survived['Sex']=='male'])
female_test=len(testset_survived[testset_survived['Sex']=='female'])
```

```python
plt.title('train set')
plt.pie([male_train,female_train],labels=['Male','Female'],autopct='%1.1f%%')
plt.plot()
```

```python
plt.title('test set')
plt.pie([male_test,female_test],labels=['Male','Female'],autopct='%1.1f%%')
plt.plot()
```
