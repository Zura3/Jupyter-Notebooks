```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
USAhousing=pd.read_csv(r'C:\Users\rohan\Downloads\USA_Housing.csv')
```

```python
USAhousing.head()
```

```python
corr=USAhousing.corr()
plt.figure(figsize=(7,7))
ax=sns.heatmap(corr,linewidths=.5,annot=True)
```

```python
X1=(USAhousing['Avg. Area Income'])
X2=(USAhousing['Area Population'])
X3=(USAhousing['Avg. Area Number of Rooms'])
Y=USAhousing['Price']
M=len(Y)
```

```python
def costFunction(theta0,theta1,theta2,theta3):
    cost=(1/(2*M))*(((theta0+theta1*X1+theta2*X2+theta3*X3-Y)**2).sum())
    return cost
```

```python
theta0,theta1,theta2,theta3,alpha=0,0,0,0,0.00000000001
ar=[]
a=1
index=[]
costArray=[]
while(a<1000):
    theta_0=theta0-alpha*(1/M)*((theta0+theta1*X1+theta2*X2+theta3*X3-Y).sum())
    theta_1=theta1-alpha*(1/M)*(((theta0+theta1*X1+theta2*X2+theta3*X3-Y)*X1).sum())
    theta_2=theta2-alpha*(1/M)*(((theta0+theta1*X1+theta2*X2+theta3*X3-Y)*X2).sum())
    theta_3=theta3-alpha*(1/M)*(((theta0+theta1*X1+theta2*X2+theta3*X3-Y)*X3).sum())
    if(costFunction(theta0,theta1,theta2,theta3)<=costFunction(theta_0,theta_1,theta_2,theta_3)):
        break
    else:
        theta0=theta_0
        theta1=theta_1
        theta2=theta_2
        theta3=theta_3
        index.append(a)
        a=a+1
        costArray.append(costFunction(theta0,theta1,theta2,theta3))
for i in range(M):
    ar.append(theta0+(theta1*X1[i])+(theta2*X2[i])+(theta3*X3[i]))
```

```python
sns.lineplot(index,costArray)
```

```python
print('theta0 = ',theta0)
print('theta0 = ',theta1)
print('theta0 = ',theta2)
print('theta0 = ',theta3)
```
